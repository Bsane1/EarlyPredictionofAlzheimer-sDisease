```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\SANEABHUTTO\Desktop\archive\archive (1)\alzheimers_disease_data.csv'
data = pd.read_csv(file_path)

# Display data types and summary statistics
print(data.info())
print(data.describe())

# Basic statistics: min, max, mean, std
statistics = data.describe().T[['min', 'max', 'mean', 'std']]
print(statistics)

# Plot histograms for all numerical attributes
data.hist(bins=20, figsize=(20, 15))
plt.show()

# Pair plot
sns.pairplot(data)
plt.show()

# Correlation heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
